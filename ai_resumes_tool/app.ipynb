```python
import os
import pdfplumber
import docx
from openai import OpenAI
from dotenv import load_dotenv


load_dotenv(override=True, dotenv_path="../.env.local")
my_api_key = os.getenv("OPENAI_API_KEY")
# my_api_key[:5], my_api_key[-10:]
client = OpenAI(api_key=my_api_key)

if my_api_key:
    print("OpenAI API Key Loaded")
else:
    raise ValueError("OPENAI_API_KEY not found")

# print("OpenAI Connected")

pdf_path = "data/ARCHANA DUBEY_PROFILE_WC.pdf"

if not os.path.exists(pdf_path):
    raise FileNotFoundError(
        f"PDF file not found: {pdf_path}. Available files: {os.listdir('data')}"
    )


def extract_text_from_pdf(path):
    text = ""
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            t = page.extract_text()
            if t:
                text += t + "\n"
    return text.strip()
def load_resume(pdf_path):
    return extract_text_from_pdf(pdf_path)
pdf_text = extract_text_from_pdf(pdf_path)
print("PDF text extracted successfully")
```

```python
resume = pdf_text

# Paste the real job description below.
job_description = """
Paste the real job description here.
"""
```

```python
def improve_resume(resume, job_description):

    prompt = f"""
You are a senior ATS resume writer helping candidates get interviews for software engineering jobs.

Your job is to rewrite and optimize the resume for the provided job description.

====================
RESUME:
{resume}

====================
JOB DESCRIPTION:
{job_description}

====================

IMPORTANT RULES:

1. Make the resume ATS optimized
2. Use strong action verbs
3. Focus on measurable impact
4. Match important keywords naturally
5. Make the candidate sound highly professional
6. Keep formatting recruiter-friendly
7. Avoid generic buzzwords
8. Keep bullet points concise and impactful
9. Do NOT invent fake experience
10. If measurable metrics are missing:
- generate realistic estimated impact examples
- make them believable and professional
- never exaggerate unrealistically
11. Make bullet points sound specific and realistic, not generic template text.
12. Tailor wording strongly to the job description.
13. Avoid repetitive sentence structures.
14.impact first approach
[Improved performance by X%]

====================

OUTPUT FORMAT:

# Professional Summary

Write a strong 3-4 line professional summary.

# Technical Skills

Categorize skills properly:
- Languages
- Frameworks
- Tools
- Cloud
- Testing

# Professional Experience

Rewrite experience using:
- strong bullet points
- action verbs
- ATS-friendly wording

# Suggested Improvements

- analyze resume
- analyze JD
- identify missing keywords
- show missing keywords
THEN rewrite

Provide:
- missing keywords
- improvement suggestions
- weak areas in resume

====================

Tone:
Professional, concise, recruiter-friendly.
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "system",
                "content": "You are a world-class resume optimization expert."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0.3
    )

    return response.choices[0].message.content
```

```python
result = improve_resume(resume, job_description)

print(result)
```
